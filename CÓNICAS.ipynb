##**Las Cónicas**

Las cónicas son una familia de curvas geométricas que se obtienen al intersectar un plano con un cono de doble hoja. Estas curvas incluyen el círculo, la elipse, la parábola y la hipérbola, cada una con características y propiedades únicas.

Las cónicas tienen muchas aplicaciones en matemáticas, física y otras áreas científicas. Son fundamentales en la descripción de órbitas planetarias, trayectorias de proyectiles, espejos parabólicos, antenas parabólicas, diseño de lentes y mucho más. Su estudio y comprensión son cruciales para comprender y analizar una variedad de fenómenos y estructuras en nuestro entorno físico y matemático.

**¿Cómo se obtienen las secciones cónicas?**

La superficie cónica se obtiene haciendo rotar una recta denominada generatriz alrededor de un punto fijo llamado vértice mantiendo otro punto constantemente sobre una circunferencia llamada directiz situaa en un plano perpendicular al eje y condicionada a que su centro esté sobre el eje. 

**Tipos de cónicas**

El círculo es una cónica especial que se forma cuando el plano de corte es perpendicular al eje del cono. Es una curva cerrada en la que todos los puntos están equidistantes del centro. Su ecuación general es (x - h)² + (y - k)² = r², donde (h, k) representa el centro del círculo y r es su radio.

La elipse se produce cuando el plano de corte es oblicuo y corta ambas hojas del cono. Es una curva cerrada y simétrica con dos ejes: el eje mayor (2a) y el eje menor (2b). Su ecuación general es ((x - h)²/a²) + ((y - k)²/b²) = 1, donde (h, k) es el centro de la elipse y a y b son las longitudes de los semiejes.

La parábola se forma cuando el plano de corte es paralelo a una de las hojas del cono. Es una curva abierta y simétrica con un eje de simetría. Su ecuación general es y = a(x - h)² + k, donde (h, k) es el vértice de la parábola y a determina la apertura y dirección de la curva.

La hipérbola ocurre cuando el plano de corte atraviesa ambas hojas del cono. Es una curva abierta con dos ramas simétricas. Tiene dos ejes: el eje transverso (2a) y el eje conjugado (2b). Su ecuación general es ((x - h)²/a²) - ((y - k)²/b²) = 1, donde (h, k) es el centro de la hipérbola y a y b determinan la forma y la orientación de la curva.

**¿Cómo se diferenciar los tipos de cónicas**

Si el discriminmante es igual a cero, entonces la cónica es una circunferencia. Si el discriminante es mayor que cero, etonces la cónica es una elipse si A y C tienen el mismo signo, o una hipérbola si tienen signos opuestos. Si el discriminante es menor que cero, entonces la cónica es una parábola.

#**Parábola**

La parábola es una cónica que se caracteriza por tener una forma curva y simétrica. Se puede definir como el lugar geométrico de todos los puntos en un plano que son equidistantes a un punto fijo llamado foco (F) y a una línea recta llamada directriz (d).

La ecuación general de una parábola en un sistema de coordenadas cartesianas es de la forma y = ax^2 + bx + c, donde a, b y c son constantes. Esta ecuación puede tener diferentes formas dependiendo de la posición y la orientación de la parábola. Algunas de las formas más comunes son:

Parábola vertical con vértice en el origen (0,0): y^2 = 4px, donde p es la distancia entre el foco y el vértice. La directriz es la línea x = -p.

Parábola vertical con vértice en (h,k): (y-k)^2 = 4p(x-h), donde (h,k) es el vértice de la parábola. La directriz es la línea x = h - p.

Parábola horizontal con vértice en el origen (0,0): x^2 = 4py, donde p es la distancia entre el foco y el vértice. La directriz es la línea y = -p.

Parábola horizontal con vértice en (h,k): (x-h)^2 = 4p(y-k), donde (h,k) es el vértice de la parábola. La directriz es la línea y = k - p.

En una parábola, el vértice es el punto más bajo o más alto de la curva, dependiendo de si la parábola se abre hacia arriba o hacia abajo. El foco y la directriz determinan la forma y la posición de la parábola. La distancia entre el foco y el vértice se denomina distancia focal (FV), y la distancia entre el vértice y la directriz se denomina distancia focal (FD), ambas están relacionadas por la ecuación 4p = FV = FD.

Las parábolas tienen numerosas aplicaciones en la física, la óptica, la ingeniería y las matemáticas aplicadas. Son utilizadas en el diseño de reflectores parabólicos, trayectorias de proyectiles, antenas parabólicas, análisis de movimiento de partículas y muchas otras áreas.

#**Hipérbola**

La hipérbola es el lugar geométrico de los puntos del plano cuya diferencia de distancias a los puntos fijos llamados focos es constante en valor absoluto.

*Elementos de la hipérbola*

Elementos de la hipérbola
 

1 Focos: Son los puntos fijos F y F'.

2 Eje focal, principal o real: Es la recta que pasa por los focos.

3 Eje secundario o imaginario: Es la mediatriz del segmento FF'.

4 Centro: Es el punto de intersección de los ejes.

5 Vértices: Los puntos A y A' son los puntos de intersección de la hipérbola con el eje focal.

6 Radios vectores: Son los segmentos que van desde un punto de la hipérbola a los focos: PF y PF'.

7 Distancia focal: Es el segmento FF' de longitud 2c.
8 Eje mayor: Es el segmento AA' de longitud 2a.

9 Eje menor: Es el segmento BB' de longitud 2b.

Los puntos B y B' se obtienen como intersección del eje imaginario con la circunferencia que tiene por centro uno de los vértices y de radio c.

10 Ejes de simetría: Son las rectas que contienen al eje real o al eje imaginario.

11 Asintotas: Son las rectas de ecuaciones: y=-b/ax,  y=b/ax

12 Relación entre los semiejes: c^2=a^2+b^2

*Excentricidad de la hipérbola*

La excentricidad es un parámetro que indica la abertura de la hipérbola. Este número, en el caso de las hipérbolas, siempre es mayor que 1.
e=c/a, c>=a, e>=1

#**Elipse**

Definición: Lugar geométrico de todos los puntos de un plano cuya suma de distancias a dos puntos fijos del mismo plano, llamados focos F y F’ es una cantidad constante, mayor que la distancia entre los focos y positiva. Esa cantidad constante es 2 a.
Elementos 
Eje focal: recta que contiene a los focos F´F
Vértices: puntos de intersección de la elipse y el eje focal. Son V y V’
Eje mayor: segmento que tiene como extremos los vértices, es decir VV’
Centro: punto medio del eje mayor. Ejemplo C
Eje normal: recta perpendicular al eje focal que pasa por el centro. NN’
Eje menor: segmento donde los extremos son los puntos de intersección del eje normal y la elipse 
Cuerda: cualquier segmento donde los extremos son dos puntos de la elipse. GG’ O AA’
Cuerda focal: cualquier cuerda que pasa por el foco JJ’ O AA’
 Lado recto: las dos cuerdas focales perpendiculares al eje focal
Diámetro: cualquier cuerda que pasa por el centro. DD’
Radio vector: cualquier cuerda que pasa por el centro.

Directrices: las rectas l y m 
Relación fundamental de la elipse es: a2=b2+c2
Donde: 
•	a: es la distancia del centro a uno de los vértices del eje mayor, es decir la mitad del eje mayor
•	 b: es la distancia del centro a uno de los vértices del eje menor, es decir la mitad del eje menor
•	 c: es la distancia del centro a uno de los focos.

```python
import matplotlib.pyplot as plt
import numpy as np

def graficar_elipse(a, b, titulo):
    # Calcular coordenadas de la elipse
    theta = np.linspace(0, 2*np.pi, 100)
    x = a * np.cos(theta)
    y = b * np.sin(theta)

    # Graficar la elipse
    plt.plot(x, y, label=titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(titulo)
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# Coordenadas del centro
centro_x = 0
centro_y = 0

# Semidistancias del eje mayor y del eje menor
a = 4
b = np.sqrt(12)

# Calcular la excentricidad
excentricidad = np.sqrt(1 - (b**2 / a**2))

# Calcular las coordenadas de los focos
c = np.sqrt(a**2 - b**2)
foco_1 = (centro_x + c, centro_y)
foco_2 = (centro_x - c, centro_y)

# Calcular las coordenadas de los vértices
vertice_1 = (centro_x + a, centro_y)
vertice_2 = (centro_x - a, centro_y)

# Mostrar la información
print(f"Focos: {foco_1}, {foco_2}")
print(f"Vértices: {vertice_1}, {vertice_2}")
print(f"Excentricidad: {excentricidad}")

# Graficar la elipse
titulo = "x^2/16 + y^2/12 = 1"
graficar_elipse(a, b, titulo)
```

#**Circunferencia**

Una circunferencia es una cónica que se define como el conjunto de todos los puntos en un plano que tienen la misma distancia al centro. Se puede describir utilizando la siguiente fórmula:

(x - h)^2 + (y - k)^2 = r^2

Donde (h, k) son las coordenadas del centro de la circunferencia, y r es el radio de la circunferencia.

En esta fórmula, (x - h) representa la diferencia entre la coordenada x de un punto y la coordenada x del centro de la circunferencia, mientras que (y - k) representa la diferencia entre la coordenada y de un punto y la coordenada y del centro de la circunferencia. El resultado de elevar al cuadrado estas diferencias y sumarlas debe ser igual al cuadrado del radio r.

En resumen, la fórmula de la circunferencia es (x - h)^2 + (y - k)^2 = r^2, donde (h, k) es el centro de la circunferencia y r es su radio. Esta fórmula permite describir la ubicación y el tamaño de una circunferencia en un plano.
